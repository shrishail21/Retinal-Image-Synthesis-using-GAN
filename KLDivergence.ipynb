```python
def display_images(image1, image2):
    fig, axes = plt.subplots(1, 2, figsize=(5, 5))
    
    axes[0].imshow(image1)
    axes[0].axis('off')
    
    axes[1].imshow(image2)
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
def prob_dist(I):
    return np.histogramdd(np.ravel(I), bins = 256)[0] / I.size
```

```python
def kl_divergence(I: np.ndarray, J: np.ndarray) -> float:
    epsilon = 1e-10
    P = prob_dist(I) + epsilon
    Q = prob_dist(J) + epsilon
    kl = np.where(P != 0, P * np.log2(P / Q), 0).sum()
    print("KL DIVERGENCE FOR REAL AND GENERATED IMAGES IS : ")
    print(kl)
```

```python
real_images = cv2.imread("/content/drive/MyDrive/DRIVE/train/manual_1st/manual_1st/manual1.png")
real_images = cv2.cvtColor(real_images, cv2.COLOR_BGR2RGB)
generated_images = cv2.imread("/content/drive/MyDrive/Vessel/Vessel/Masked_Vessel01.png")
generated_images = cv2.cvtColor(generated_images, cv2.COLOR_BGR2RGB)
real = cv2.imread("/content/drive/MyDrive/DRIVE/train/manual_1st/manual_1st/manual1.png")
generated = cv2.imread("/content/drive/MyDrive/Vessel/Vessel/Masked_Vessel01.png")
kl_divergence(real_images, generated_images)
display_images(real, generated)
```
